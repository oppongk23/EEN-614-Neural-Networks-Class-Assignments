```python
#Question 2
"""Write a function to generate and print the first (n+1) Fibonacci numbers. 
The main program should ask the user to enter the n and then pass it on to the function.
"""
def fibGenerator(length):
    """This program generates the Fibonacci number sequence of length N + 1 where N is provided by the user"""
    #This line flags an error if the input provided is negative
    if length < 0:  
        raise Exception("The value of the length (n) can only be a positive number")
    #The fibonacci array is initialized to these [0,1] to initialize in case the input is either 0 or 1
    fib_array = [0, 1]
    
    if length == 0:
        return
    
    if length == 1:
        print(*fib_array)
        return
    
    while len(fib_array) <= length:
            #the next number in the sequence is the sum of the two most recent values. The value is then appended to the sequence
            next_number = fib_array[-1] + fib_array[-2]
            fib_array.append(next_number)
    print(*fib_array)

def main():
    n = int(input("Please enter the value of the length 'N' of the Fibonacci Sequence: "))
    fibGenerator(n)
    
main()
```

```python
#Question 3
"""Write a program to generate a random 20x3 matrix representing Cartesian coordinates in a 
three dimensional space and then convert them into spherical coordinates.
"""
#the sperical coordinates are written in the form (rho, theta, phi)
random_cartesian_coordinates = np.random.rand(20,3)
def convertor(x, y, z):
    """This convertor returns rho, theta and phi with theta and phi being in radians"""
    #calculate the values of rho, theta, and phi
    rho = np.sqrt((x ** 2) + (y ** 2) + (z ** 2))
    theta = np.arctan(y/x)
    phi = np.arccos(z/rho)
    
    return [rho, theta, phi]

random_spherical_coordinates = np.array([convertor(x, y, z) for x, y, z in random_cartesian_coordinates])

print("These are the cartesian coordinates", random_cartesian_coordinates)
print("\n")
print("These are the spherical coordinates", random_spherical_coordinates)
    
    
```
